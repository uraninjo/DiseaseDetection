```python
import pandas as pd
import numpy as np
import seaborn as sbn
import matplotlib.pyplot as plt

df=pd.read_csv('Parkinsson disease.csv')
df
```

```python
df.drop('name',axis=1,inplace=True)
```

```python
df.columns
```

```python
df.describe().T
```

```python
df['MDVP(Hz)']=(df['MDVP:Fo(Hz)']+df['MDVP:Fhi(Hz)']+ df['MDVP:Flo(Hz)'])/3
df['NHR/HNR']=(df['NHR']*df['HNR'])
df['RPDE/PPE']=(df['RPDE']+df['PPE'])/2
df['spread1/spread2']=df['spread1']*df.spread2
```

```python
df.drop(['MDVP:Fo(Hz)', 'MDVP:Fhi(Hz)', 'MDVP:Flo(Hz)'],axis=1,inplace=True)
df.drop(['NHR', 'HNR', 'RPDE','PPE','spread1','spread2'],axis=1,inplace=True)
```

```python
X_train = df.drop("status", axis=1)
y_train = df.status

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.model_selection import cross_validate, GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.linear_model import SGDClassifier
from sklearn.naive_bayes import GaussianNB
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier,StackingClassifier


from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import classification_report, accuracy_score, f1_score, confusion_matrix
```

```python
classifiers = {
    'Catboost': CatBoostClassifier(),
    "NB": GaussianNB(),
    "SGD": SGDClassifier(),
    "KNN": KNeighborsClassifier(), 
    "LR": LogisticRegression(max_iter=1000), 
    "DT": DecisionTreeClassifier(),
    "RF": RandomForestClassifier(n_estimators=100, max_depth=3, random_state=2),
    "SVM": SVC(),
    "MLP": MLPClassifier(max_iter=1000),
    "XGB": XGBClassifier(),
    "LGBM": LGBMClassifier(),
    'GradB':GradientBoostingClassifier()
}


results = pd.DataFrame(columns=["Classifier", "Avg_Accuracy", "Avg_F1_Score"])
for name, clf in classifiers.items():
    model = clf
    cv_results = cross_validate(
        model, X_test_scaled, y_test, cv=10,
        scoring=(['accuracy', 'f1'])
    )

    results = results.append({
        "Classifier": name,
        "Avg_Accuracy": cv_results['test_accuracy'].mean(),
        "Avg_F1_Score": cv_results['test_f1'].mean()
    }, ignore_index=True)
    
results["Avg_Overall"] = (results["Avg_Accuracy"] + results["Avg_F1_Score"]) / 2
results = results.sort_values("Avg_Overall", ascending=False)
results
```

```python
svm = XGBClassifier()
params = {
    'learning_rate':[i*0.001 for i in range(1,101,5)],
    'max_depth':[i for i in range(1,10)],
    "n_estimators": [i for i in range(0,200,50)]
}
clf = GridSearchCV(svm, params, cv=10)
clf.fit(X_train_scaled, y_train)
print("Best hyperparameter:", clf.best_params_)
```

```python
results=[]
cv_results = cross_validate(
        svm, X_test_scaled, y_test, cv=10,
        scoring=(['accuracy', 'f1'])
    )

pd.DataFrame([cv_results['test_accuracy'].mean(),cv_results['test_f1'].mean()])
```

```python
def scores(y_test,y_pred):
    from sklearn.metrics import confusion_matrix,precision_score,recall_score,f1_score
    cm=confusion_matrix(y_test,y_pred)
    ps=precision_score(y_test,y_pred),
    rs=recall_score(y_test,y_pred)
    f1=f1_score(y_test,y_pred)
    print(cm)
    df=pd.DataFrame([['precision_score',ps[0]],
    ['recall_score',rs],['f1',f1]],columns=['Metrics','Scores'])
    return df


y_pred=clf.predict(X_test_scaled)
scores(y_test,y_pred)
```
